```python
def uso_maximo(lista):
  df = pd.DataFrame(lista)
  data = []
  indexFor = df.shape
  for i in range(indexFor[0]):
    max = df.iloc[i].max()
    data.append(max)
  return print(data)
```

```python
uso_maximo(uso_cpu)
```

```python
def media(lista):
  data = 0
  media = 0
  df = pd.DataFrame(lista)
  indexFor = df.shape
  for i in range(indexFor[1]):
    data = 0
    media = int(df[f'Região {i+1}'].mean())
    for h in range(indexFor[0]):
      if df.iloc[h][f'Região {i+1}'] > media:
        data += 1
    print(f'media: {media} quantidade: {data}')
```
